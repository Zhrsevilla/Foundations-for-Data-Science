```python
df_e['BMI'] = df_e['berat_badan'] / (df_e['tinggi_badan'] / 100) ** 2
df_e
```

```python
def hitung_bmi(berat_badan, tinggi_badan):
    bmi = berat_badan / ((tinggi_badan/ 100) ** 2)
    return bmi

# Input berat (dalam kilogram) dan tinggi (dalam meter) dari pengguna
berat = float(input("Masukkan berat Anda (kg): "))
tinggi = float(input("Masukkan tinggi Anda (cm): "))

# Menghitung BMI menggunakan fungsi hitung_bmi
bmi = hitung_bmi(berat, tinggi)

# Menampilkan hasil BMI
print("BMI Adam Ibnudzaky Handoko :", bmi)
```

```python
# Menentukan kategori BMI untuk setiap siswa

def kategori_bmi(bmi):
  if bmi < 18.5:
    print("Underweight")
  elif 18.5 <= bmi < 24.9:
    print("Normal")
  elif 24.9 <= bmi < 29.9:
    print("Overweight")
  else:
    print("Obesity")

df_e["Status BMI"] = df_e["BMI"].apply(kategori_bmi)
```

```python
# Mengurutkan DataFrame berdasarkan tinggi badan dari tertinggi ke terendah
df_e1 = df_e.sort_values(by='tinggi_badan', ascending=False)

df_e1
```
